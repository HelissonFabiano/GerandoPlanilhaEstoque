**Gerando uma planilha para controle de estoque**

Com a intenção de ajudar uma revendedora de cosméticos a administrar seus produtos, resolvi gerar uma planilha no Excel para poder controlar informações como: nome do produto, código, preço de compra, preço de venda, etc... Durante o desenvolvimento dessa planilha, cheguei a conclusão de que digitar manualmente as informações de cada produto seria muito trabalhoso, pela quantidade de produtos existentes.
Atuando como Cientista de Dados, uma dúvida me veio à cabeça: como eu poderia optimizar o tempo para a criação dessa planilha?
Foi então que tive a seguinte ideia: A empresa que fornece os cosméticos geram um PDF com as informações detalhadas da compra feita pela revendedora, e eu poderia utilizar esse PDF como entrada de dados para a minha planilha.
